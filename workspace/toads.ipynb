```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cols = ["customer_id", "merchant_id", "n_toads_2014", "n_subs_2014", "merchant_id_2015","n_toads_2015", "n_subs_2015"]
toads = pd.read_csv("data/toads.csv", names=cols)
```

```python
plt.figure(figsize=(20,10))
rp = sns.regplot(x="n_toads_2014", y="n_toads_2015", data=toads, x_estimator=np.mean)
#             x_jitter=0.5, y_jitter=0.5, 
#             color="#000000", marker='.', scatter_kws={"alpha":0.5})

rp.set(xlim=[0, 30], ylim=[0, 20])
```

```python
toads["toads_perc_2014"] = toads["n_toads_2014"] / toads["n_subs_2014"]
toads["toads_perc_2015"] = toads["n_toads_2015"] / toads["n_subs_2015"]
toads_nontrivial = toads[toads["toads_perc_2015"]>0][toads["toads_perc_2015"]<1][toads["toads_perc_2014"]>0][toads["toads_perc_2014"]<1]
```
